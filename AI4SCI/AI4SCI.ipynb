```python
import pandas as pd
import glob
import os

# 指定CSV文件所在的目录
input_folder = "C:\\Users\\30396\\PycharmProjects\\pythonProject\\AI4SCI\\stock data"   # 替换为你实际的文件夹路径
output_folder = 'C:\\Users\\30396\\PycharmProjects\\pythonProject\\AI4SCI\\clean stock data'  # 保存清洗后文件的文件夹

# 如果输出文件夹不存在，创建一个新的
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
    print("Created")

# 获取目录下所有的CSV文件
csv_files = glob.glob(os.path.join(input_folder, '*.csv'))

# 需要保留的列（参考你的截图中列出的指标）
columns_to_keep = [
    'ts_code','trade_date', 'open_qfq', 'close_qfq', 'pct_chg', 'vol', 'amount', 'turnover_rate', 
    'adj_factor', 'atr_bfq', 'atr_qfq', 'brar_ar_bfq', 'brar_ar_qfq', 
    'brar_br_bfq', 'brar_br_qfq', 'psy_bfq', 'psy_qfq', 'psyma_bfq', 'psyma_qfq'
]

# 批量处理每个CSV文件
for file_path in csv_files:
    try:
        # 读取CSV文件
        df = pd.read_csv(file_path)
        
        # 保留需要的列
        df = df[columns_to_keep]
        
        # 处理缺失值（可以根据需求修改为不同的处理方法）
        # 例如填充缺失值、删除缺失值行等
        df = df.dropna()  # 删除含有NaN的行

        # 生成清洗后的文件路径
        file_name = os.path.basename(file_path)
        output_path = os.path.join(output_folder, file_name)

        # 保存清洗后的文件
        df.to_csv(output_path, index=False)
        print(f"清洗完成: {output_path}")
        
    except Exception as e:
        print(f"处理文件 {file_path} 时出错: {e}")
        
print("所有文件清洗完成！")
```
